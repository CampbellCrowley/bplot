```python
import numpy as np
import bplot as bp
import scipy.stats as stats
```

## Box

```python
N = 1001
y = np.random.normal(size=N)
x = np.arange(N)
```

```python
bp.box(0, y)
bp.title("A box plot")
bp.tight_layout()
```

```python
bp.box_h(y, 0, color='orange', label='Yep, a label')
bp.legend()
```

## Curve

```python
bp.curve(x, y, style='--')
```

## Density

```python
bp.density(y, alpha=0.25, color='green')
```

## Histogram

```python
bp.histogram(y, alpha=0.25, color='grey')
```

```python
bp.histogram(y.tolist(), bins='bb')
```

## Jitter

```python
bp.jitter(x, y)
```

```python
bp.jitter([1,2], [3,4], jitter_x = 0.1, jitter_y = 0.1)
```

## Lag

```python
bp.lag(y)
```

## Line

```python
bp.line(x, y, style='-.', size=0.5)
bp.line_h(0, xmin=0, xmax=1000, color='purple', size=2)
bp.line_v(200, -2, 2, color='black', size=2)
```

## Letter Value

```python
bp.lv(0, y, color='green')
```

## Median Absolute Deviation interval

```python
bp.mad(0, y)
```

```python
bp.mad_h(y, 0)
```

## Percentile interval

At scalar $x$, plot vertically the median as the point and 50 and 80% intervals as calculated from the 25th and 75th and 10th and 90th percentiles, respectively.

```python
bp.percentile(0, y)
```

At scalar $y$, plot horizontally the median as the point and 50 and 80% intervals as calculated from the 25th and 75th and 10th and 90th percentiles, respectively.

```python
bp.percentile_h(y, 0)
```

## Rug

```python
bp.rug(y)
```

## Scatter

```python
bp.scatter(x, y)
```

## Shade

```python
z = np.linspace(-3, 3, 101)
f = stats.norm.pdf(z)
bp.curve(z, f)

sx = np.linspace(-2, -1, 11)
bp.shade_x(sx, np.repeat(0, sx.size), stats.norm.pdf(sx))
```

```python
g = z**2
bp.curve(z, g)

sy = np.linspace(2, 4, 11)
bp.shade_y(sy, np.repeat(-3, sy.size), -np.sqrt(sy))
```

## Standard Deviation interval

```python
bp.std(0, y)
```

```python
bp.std_h(y, 0)
```

## Standard Error interval

```python
bp.ste(0, y)
```

```python
bp.ste_h(y, 0)
```

## Trace

```python
bp.trace(y)
```
